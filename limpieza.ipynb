```python
import ast
import pandas as pd
```

```python
rows = []
with open('steam_games.json') as f:
    for line in f.readlines():
        rows.append(ast.literal_eval(line))
        
df = pd.DataFrame(rows)

# Convertir la columna "release_date" al tipo datetime
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
```

```python
# Obtener la frecuencia de cada valor en la columna 'app_name'
frecuencia_valores = df['app_name'].value_counts()

# Filtrar los valores que se repiten en la columna 'app_name'
valores_repetidos = frecuencia_valores[frecuencia_valores > 1].index.tolist()

# Obtener las filas con los valores repetidos
filas_valores_repetidos = df[df['app_name'].isin(valores_repetidos)]
filas_valores_repetidos
```

```python
# Filtrar las filas donde 'app_name' es NaN
rows_with_nan = df[df['app_name'].isnull()]

# Imprimir las filas con valores NaN en 'app_name'
rows_with_nan
```

En el DataFrame, se han identificado dos registros con información incompleta. Uno de ellos contiene valores en la columna 'specs' pero carece de información en la columna 'precio'. Por otro lado, el segundo registro tiene información en la columna 'precio', pero no presenta valores en la columna 'specs'. Dado esto decido eliminarlos

```python
# Identificar las filas duplicadas en la columna
filas_duplicadas = df.duplicated(subset=['app_name'], keep=False)

# Eliminar las filas que contienen valores duplicados en la columna
df = df[~filas_duplicadas]
```

```python
# Mostrar si tuvo efecto
frecuencia_valores = df['app_name'].value_counts()
valores_repetidos = frecuencia_valores[frecuencia_valores > 1].index.tolist()
print(valores_repetidos)
```

```python
# Mostrar si tuvo efecto
rows_with_nan = df[df['app_name'].isnull()]
rows_with_nan
```

```python
# Guardar el DataFrame limpio en un archivo csv
df.to_csv('steam_games_limpio.csv')
```
